```python
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

recipe = df.columns[1:]

data = [100, 250, 200, 120, 80]

def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d})".format(pct, absolute)
```

```python
wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, recipe,
          title="Asset",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Assets")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Criar o DataFrame
data = {
    'Date': ['2023-07-10', '2023-07-11', '2023-07-12', '2023-07-10'],
    'Asset': ['Bond B', 'Bond A', 'Bond B', 'Bond A'],
    'Price': [385.64, 299.99, 385.64, 299.99]
}
df = pd.DataFrame(data)

# Contar as ocorrências de cada item na coluna 'Asset'
asset_counts = df['Asset'].value_counts()

# Criar o gráfico de pizza
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

wedges, texts, autotexts = ax.pie(asset_counts, autopct=lambda pct: func(pct, asset_counts),
                                  textprops=dict(color="w"))

ax.legend(wedges, asset_counts.index,
          title="Asset",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Assets")

plt.show()
```
